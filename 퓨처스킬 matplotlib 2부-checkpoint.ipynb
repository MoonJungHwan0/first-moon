```python
#5

X = ["Tom", "Amy", "Jim", "Bill", "Pegi"]
Y = [8, 23, 52, 37, 64]


# bar
# 바 그래프의 색을 "lightgray", 테두리 색을 "gray"로 지정합니다.

fig, ax= plt.subplots()
ax.bar(X,Y,color='lightgray',edgecolor="gray") # edgecolor="gray" 추가


# label
# X, Y축에 각기 "Salesperson"과 "Revenue (1000 $)" 라는 레이블을 달아줍니다.
# 폰트 크기는 14, 두껍게("bold"), 회색("gray")으로 씁니다.
# X, Y 축과의 거리는 각각 12만큼 띄웁니다.

ax.set_xlabel('Salesperson',size=14,color='gray',weight='bold',labelpad=12) # labelpad 축과의 거리 띄우는거 추가
ax.set_ylabel('Revenue(1000$)',size=14,color='gray',weight='bold',labelpad=12)

plt.show()

plt.savefig('abcde')
```

```python
#6
X = ["Tom", "Amy", "Jim", "Bill", "Pegi"]
Y = [8, 23, 52, 37, 64]


fig, ax= plt.subplots()
ax.bar(X,Y,color='lightgray',edgecolor="gray")

ax.set_xlabel('Salesperson',size=14,color='gray',weight='bold',labelpad=12) 
ax.set_ylabel('Revenue(1000$)',size=14,color='gray',weight='bold',labelpad=12)


for i, v in enumerate(X):
    ax.text(v, Y[i], Y[i], # 좌표 (x축 = v, y축 = y[0]..y[1], 표시 = y[0]..y[1])
            fontsize = 9,
            horizontalalignment='center', # horizontalalignment (left, center, right) / 수평정렬
            verticalalignment='bottom') # verticalalignment (top, center, bottom) / 수직 정렬
    
plt.show()

plt.savefig('text a')

#참조 : https://pydole.tistory.com/entry/Python-matplotlib-%EA%B7%B8%EB%9E%98%ED%94%84%EC%97%90-%EA%B0%92-%ED%91%9C%EC%8B%9C-%ED%95%98%EA%B8%B0
```

```python
#7
X = ["Tom", "Amy", "Jim", "Bill", "Pegi"]
Y = [8, 23, 52, 37, 64]


fig, ax= plt.subplots()
ax.bar(X,Y,color='lightgray',edgecolor="gray",zorder=2) #zorder 함수 추가



ax.set_xlabel('Salesperson',size=14,color='gray',weight='bold',labelpad=12) 
ax.set_ylabel('Revenue(1000$)',size=14,color='gray',weight='bold',labelpad=12)

ax.axhline(y=45,color='blue',linestyle='--',zorder=1) #zorder 함수 추가
ax.axhline(y=15,color='r',linestyle='--',zorder=1)

ax.axhspan(0,15,color='r',alpha=0.2,zorder=0 ) # 음영추가


for i, v in enumerate(X):
    ax.text(v, Y[i], Y[i], # 좌표 (x축 = v, y축 = y[0]..y[1], 표시 = y[0]..y[1])
            fontsize = 9,
            horizontalalignment='center', # horizontalalignment (left, center, right) / 수평정렬
            verticalalignment='bottom') # verticalalignment (top, center, bottom) / 수직 정렬
    

    
plt.show()
plt.savefig('axhline')

```

```python
ax.fill_between(X,0,15,alpha=0.2,color='r' #음영추가 다른코드 그런데 X값에만 음영이 채워짐

#7번 코드 추가 설명
bar graph의 zorder = 2, hline의 zorder = 1 로 설정해주니 막대그래프 뒤로 직선이 지나가게 그려졌습니다
기본적으로 axes에 그려지는 순서는 patches(다양한 도형 컬렉션), 직선, 글자 순서 인데 
zorder value에 따라 다르게 그려집니다. 가장 바깥에 그려질수록 zorder가 높아지면 됩니다. 
개인적인 생각으로는 x,y 평면에 z축이 꽂혀서 그 위에 레이어를 쌓아가면서 그린다는 느낌입니다
```

```python
#8
X = ["Tom", "Amy", "Jim", "Bill", "Pegi"]
Y = [8, 23, 52, 37, 64]


fig, ax= plt.subplots()
ax.bar(X,Y,color='lightgray',edgecolor="gray",zorder=2) #zorder 함수 추가



ax.set_xlabel('Salesperson',size=14,color='gray',weight='bold',labelpad=12) 
ax.set_ylabel('Revenue(1000$)',size=14,color='gray',weight='bold',labelpad=12)

ax.axhline(y=45,color='blue',linestyle='--',zorder=1) #zorder 함수 추가
ax.axhline(y=15,color='r',linestyle='--',zorder=1)

ax.axhspan(0,15,color='r',alpha=0.2,zorder=0 ) # 음영추가

ax.grid(True,axis='y',zorder=0) #그리드 설정

ax.spines['right'].set_visible(False) #오른쪽 테두리 제거
ax.spines['top'].set_visible(False) #위쪽 테두리 제거
ax.spines['left'].set_visible(False)#왼쪽 테두리 제거
ax.spines['bottom'].set_linewidth(2)# 테두리를 두껍게

#테두리 제거 함수 참조 : https://velog.io/@khnn/TIL-Matplotlib%EC%9C%BC%EB%A1%9C-%EC%B0%A8%ED%8A%B8-%EA%BE%B8%EB%AF%B8%EA%B8%B0/ 테두리 없애기 부분
for i, v in enumerate(X):
    ax.text(v, Y[i], Y[i], # 좌표 (x축 = v, y축 = y[0]..y[1], 표시 = y[0]..y[1])
            fontsize = 9,
            horizontalalignment='center', # horizontalalignment (left, center, right) / 수평정렬
            verticalalignment='bottom') # verticalalignment (top, center, bottom) / 수직 정렬
    

    
plt.show()
plt.savefig('spines grid')
```

```python
#9
X = ["Tom", "Amy", "Jim", "Bill", "Pegi"]
Y = [8, 23, 52, 37, 64]


fig, ax= plt.subplots()
ax.bar(X,Y,color='lightgray',edgecolor="gray",zorder=2) #zorder 함수 추가



ax.set_xlabel('Salesperson',size=14,color='gray',weight='bold',labelpad=12) 
ax.set_ylabel('Revenue(1000$)',size=14,color='gray',weight='bold',labelpad=12)

ax.axhline(y=45,color='blue',linestyle='--',zorder=1) #zorder 함수 추가
ax.axhline(y=15,color='r',linestyle='--',zorder=1)

ax.axhspan(0,15,color='r',alpha=0.2,zorder=0 ) # 음영추가

ax.grid(True,axis='y',zorder=0) #그리드 설정

ax.spines['right'].set_visible(False) #오른쪽 테두리 제거
ax.spines['top'].set_visible(False) #위쪽 테두리 제거
ax.spines['left'].set_visible(False)#왼쪽 테두리 제거
ax.spines['bottom'].set_linewidth(2)# 테두리를 두껍게


ax.yaxis.set_major_locator(plt.MultipleLocator(15)) #간격을 15로 조정
ax.tick_params(axis = 'x', length = 0)
ax.tick_params(axis = 'y', length = 0, labelsize = 'large')

#테두리 제거 함수 참조 : https://velog.io/@khnn/TIL-Matplotlib%EC%9C%BC%EB%A1%9C-%EC%B0%A8%ED%8A%B8-%EA%BE%B8%EB%AF%B8%EA%B8%B0/ 테두리 없애기 부분
for i, v in enumerate(X):
    ax.text(v, Y[i], Y[i], # 좌표 (x축 = v, y축 = y[0]..y[1], 표시 = y[0]..y[1])
            fontsize = 9,
            horizontalalignment='center', # horizontalalignment (left, center, right) / 수평정렬
            verticalalignment='bottom') # verticalalignment (top, center, bottom) / 수직 정렬
    

    
plt.show()
plt.savefig('ticks')
```

```python
#다른코드
def line():
  print("-"*10)
col = list('ABC')
A = pd.DataFrame([[1,3,5],[2,4,6]],columns=col)
B = pd.DataFrame([[-1,-3,-5],[-2,-4,-6]],columns=col)
line()
print("DataFrame A")
print(A)
line()
print("DataFrame B")
print(B)
line()
# axis는 어느축으로 합칠건지 정한다
# 0은 가로축을 기준으로 합친다 위,아래
# 1은 세로축을 기준으로 합친다 오른쪽,왼쪽
# ignore_index는 합칠 인덱스를 무시할건지 여부를 정한다
# True를 할경우 합칠 데이터 프레임의 인덱스를 무시하고 
# False를 할경우 합칠 데이터 프레임의 인덱스를 보존한상태로 데이터프레임을 만든다
C1 = pd.concat([A,B],axis = 0,ignore_index=True)
print(C1)
line()
# concat과 기능은 비슷하지만 append는 concat처럼 여러개를 합치는게 불가능하다
# axis 값을 줄수 없다 기능도 concat보다 적다
C2 = A.append(B,ignore_index=True)
print(C2)
line()
# merge는 데이터베이스에서 join 을 반영한 병합방식
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html?highlight=merge#pandas.DataFrame.merge
# 내용이 많기때문에 공식문서를 참고하면 될것같다
C3 = pd.merge(A,B,how="outer")
print(C3)
line()
```

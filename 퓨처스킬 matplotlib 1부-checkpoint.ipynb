```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#1
X = [0, 1, 2, 3, 4]
Y = [10, 20, 50, 40, 60]
plt.plot(X,Y)
plt.show()

plt.savefig('파일 이름')
```

```python
#2
X = [0, 1, 2, 3, 4]
Y = [10, 20, 50, 40, 60]
title = "Revenue (1000 $)"

plt.bar(X,Y)
plt.title('Revenue(1000$)')
plt.show()

plt.savefig('BAR')
```

```python
#3

penguins = sns.load_dataset("penguins")
# seaborn에 내장된 펭귄 체형 데이터를 불러옵니다.

X = penguins["flipper_length_mm"]
# 펭귄 날개 길이를 X 변수로 지정합니다.

Y = penguins["body_mass_g"]
# 펭귄 체중을 Y 변수로 지정합니다.

title = "Flipper length(mm) vs Body mass(g)"
# 날개 길이와 체중의 상관관계를 그립니다.

plt.scatter(X,Y)
plt.title(title)
plt.xticks(np.arange(160,250,10)) # x축 눈금 설정
plt.yticks(np.arange(2500,7000,500))# y축 눈금 설정
plt.show()

plt.savefig('Scatter')
```

```python
#다른코드, x축,y축 눈금 설정
plt.xlim([160,240])
plt.ylim([2500,6500]) 



```

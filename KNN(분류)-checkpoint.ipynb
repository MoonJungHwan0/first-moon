```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')

print(df.shape)
df.head()
```

```python
df.info()
```

```python
df1 = df.drop(['deck','embark_town'],axis=1) #컬럼 삭제
print(df1.columns.values)
print(df1.shape)
df1.head()
```

```python
df2 = df1.dropna(subset=['age'],how='any',axis=0)
print(df2.shape)
df2.head()
```

```python
df2.embarked.value_counts(dropna=True)
```

```python
df2.embarked.value_counts(dropna=True).idxmax() #가장 많은값을 찾아줌
```

```python
 df2.describe(include='all') #통계치
```

```python
df2.embarked.isnull().sum()
```

```python
df2['embarked']. fillna('S',inplace=True) #결측값을 s로 채움
```

```python
df2.embarked.isnull().sum()
```

```python
df2.columns
```

```python
df3 = df2[['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch','embarked',]]
df3.head()
```

```python
onehot_sex = pd.get_dummies(df3['sex'])
onehot_sex
```

```python
df4 = pd.concat([df3,onehot_sex],axis=1) # 옆으로 합침
df4.head()
```

```python
onehot_embarked = pd.get_dummies(df4['embarked'],prefix='town')

df5= pd.concat([df4,onehot_embarked],axis=1)

df5.head()
```

```python
df5.columns
```

```python
X= df5[['pclass', 'age', 'sibsp', 'parch', 'female', 'male', 'town_C', 'town_Q', 'town_S']]

y= df5['survived']

from sklearn import preprocessing
X= preprocessing.StandardScaler().fit(X).transform(X)
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test , y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=10)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier


knn= KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train,y_train)

y_hat = knn.predict(X_test)

print(y_hat[0:10])
print(y_test.values[0:10])
```

```python
from sklearn import metrics

knn_matrix = metrics.confusion_matrix(y_test,y_hat)
print(knn_matrix)
# 215명중 에서 미생존자가 109명, 미생존자를 생존자를 분류한 fp가 16명 
#생존자를 미생존자라고 예측한 사람 fn은 25 생존자를 정확하게 예측한 tn은 65명
```

```python
knn_report = metrics.classification_report(y_test,y_hat)
print(knn_report)
```

# Dicision_Tree

```python
uci_path ='http://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data'
df= pd.read_csv(uci_path,header=None)

print(df.shape)
df.head()
```

```python
df.columns=['id','clump','cell_size','cell_shape','adhesion','epithlial',
            'bare_nuclei','chromatin','normal_nucleoli','mitoses','class']
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['bare_nuclei'].unique()
```

```python
df.bare_nuclei.replace('?',np.nan,inplace=True) # ?를 넌값으로 변경하는 코드
```

```python
df['bare_nuclei'].unique()
```

```python
df.dropna(subset=['bare_nuclei'],axis=0,inplace=True) # 누락데이터 날리는 코드 
df.bare_nuclei = df.bare_nuclei.astype('int')

df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
X=df[['clump', 'cell_size', 'cell_shape', 'adhesion', 'epithlial','bare_nuclei', 'chromatin', 'normal_nucleoli', 'mitoses']]

y=df['class']

X=preprocessing.StandardScaler().fit(X).transform(X) #정규화
print(X)
```

```python
X_train, X_test , y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=10)

```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn import tree 
```

```python
# 모형학습

tree_model = tree.DecisionTreeClassifier(criterion='entropy',max_depth=5)

tree_model.fit(X_train,y_train)

y_hat = tree_model.predict(X_test) # predict 예측

print(y_hat[0:10])
print(y_test.values[0:10])
```

```python
from sklearn import metrics

tree_matrix = metrics.confusion_matrix(y_hat,y_test)

print(tree_matrix)
```

```python
from sklearn import metrics
tree_report = metrics.classification_report(y_test,y_hat)
print(tree_report)
```
